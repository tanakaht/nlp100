Wikipediaの記事を以下のフォーマットで書き出したファイルjawiki-country.json.gzがある．

1行に1記事の情報がJSON形式で格納される
各行には記事名が"title"キーに，記事本文が"text"キーの辞書オブジェクトに格納され，そのオブジェクトがJSON形式で書き出される
ファイル全体はgzipで圧縮される
以下の処理を行うプログラムを作成せよ．


````python
import json
from pprint import pprint


wikipath = 'data/03-re/jawiki-country.json'
````

20. JSONデータの読み込み
Wikipedia記事のJSONファイルを読み込み，「イギリス」に関する記事本文を表示せよ．問題21-29では，ここで抽出した記事本文に対して実行せよ．


````python
text = ''
with open(wikipath, 'r') as wiki:
    for line in wiki:
        data = json.loads(line)
        if data['title'] == 'イギリス':
            print(data['text'])
            text += data['text'] + '\n'

ukpath = 'data/03-re/uk.txt'
with open(ukpath, 'w') as f:
    f.write(text[:-1])
````

21. カテゴリ名を含む行を抽出
記事中でカテゴリ名を宣言している行を抽出せよ．


````python
import re

pattern = re.compile('^\[\[Category:.*?\]\]', re.MULTILINE)
with open(ukpath, 'r') as f:
    categorys = re.findall(pattern, f.read())
for category in categorys:
    print(category)
````

22. カテゴリ名の抽出
記事のカテゴリ名を（行単位ではなく名前で）抽出せよ．


````python
pattern = re.compile('^(\[\[Category:)(.*?)(\|.*?)?\]\]', re.MULTILINE)
with open(ukpath, 'r') as f:
    categorys = re.findall(pattern, f.read())
for category in categorys:
    print(category[1])
````

23. セクション構造
記事中に含まれるセクション名とそのレベル（例えば"== セクション名 =="なら1）を表示せよ．


````python
pattern = re.compile('^(===*)(.*?)(===*)', re.MULTILINE)
with open(ukpath, 'r') as f:
    sections = re.findall(pattern, f.read())
for section in sections:
    print(f'{section[1]} level={len(section[0])-1}')
````

24. ファイル参照の抽出
記事から参照されているメディアファイルをすべて抜き出せ．


参照パターン

- [[File:Battle of Waterloo 1815.PNG|thumb|left|[[ワーテルローの戦い]]
- [[ファイル:CHANDOS3.jpg|thumb|150px|[[ウィリアム・シェイクスピア]]]]
- <gallery>
ファイル:PalaceOfWestminsterAtNight.jpg|[[ウェストミンスター宮殿]]
</gallery>

````python
pattern = re.compile('\[\[(File|ファイル):(.*?)\|')
with open(ukpath, 'r') as f:
    files = re.findall(pattern, f.read())
for file in files:
    print(f'{file[1]}')

gallerypattern = re.compile('\<gallery\>(.*?)\</gallery\>', re.MULTILINE | re.DOTALL)
gallery2file = re.compile('^(File|ファイル):(.*?)\|', re.MULTILINE)
with open(ukpath, 'r') as f:
    galleryfiles = re.findall(gallerypattern, f.read())
for galleryfile in galleryfiles:
    for file in re.findall(gallery2file, galleryfile):
        print(f'{file[1]}')
````

25. テンプレートの抽出
記事中に含まれる「基礎情報」テンプレートのフィールド名と値を抽出し，辞書オブジェクトとして格納せよ．


````python
kiso_pattern = re.compile('^\{\{基礎情報(.*?)^\}\}', re.MULTILINE | re.DOTALL)
with open(ukpath, 'r') as f:
    kiso = re.search(kiso_pattern, f.read()).group()
tmpl_pattern = re.compile('^\|(.*?) = (.*?)$', re.MULTILINE)
tmpl_dict = {key: value for key, value in re.findall(tmpl_pattern, kiso)}
pprint(tmpl_dict)
````

26. 強調マークアップの除去
25の処理時に，テンプレートの値からMediaWikiの強調マークアップ（弱い強調，強調，強い強調のすべて）を除去してテキストに変換せよ（参考: マークアップ早見表）．


````python
emph_pattern = re.compile("\'{2,5}")
def rm_emph(text):
     return re.sub(emph_pattern, "", text)

tmpl_dict_no_emph = {key: value for key, value in re.findall(tmpl_pattern, rm_emph(kiso))}
pprint(tmpl_dict_no_emph)

````

27. 内部リンクの除去
26の処理に加えて，テンプレートの値からMediaWikiの内部リンクマークアップを除去し，テキストに変換せよ（参考: マークアップ早見表）．


|変換前|変換後|
|:--|:--|
|[[記事名]]|記事名|
|[[記事名\|表示文字]]|表示文字|
|[[記事名#節名\|表示文字]]|表示文字|

````python
inside_link_pattern = re.compile("\[\[([^|]*?\|)??([^|]*?)\]\]")
def rm_inside_link(text):
    return re.sub(inside_link_pattern, '\\2', text)

tmpl_dict_no_emph_link = {key: value for key, value in re.findall(tmpl_pattern, rm_inside_link(rm_emph(kiso)))}
pprint(tmpl_dict_no_emph_link)

````

28. MediaWikiマークアップの除去
27の処理に加えて，テンプレートの値からMediaWikiマークアップを可能な限り除去し，国の基本情報を整形せよ．


```
<ref name="hogehoge" />
<ref name="hogehoge"> hogehoge </ref>
<ref> hogeehoge </ref>
<ref> hogehoge \n
<br/>
<br />
{{lang|sco|hoge}}
(&pound;)
 <references /> (=ref?)
```

````python
import html

ref_ptn = re.compile('<ref.+?(\/>|<\/ref>)', re.DOTALL)
def rm_ref(text):
    return re.sub(ref_ptn, '', text)

br_ptn = re.compile('<br\s?\/>')
def rm_br(text):
    return re.sub(br_ptn, '', text)

lang_ptn = re.compile('\{\{[^\|]*?\|[^\|]*?\|([^\|]*?)\}\}')
def rm_lang(text):
    return re.sub(lang_ptn, '\\1', text)
    
good_text = html.unescape(rm_lang(rm_br(rm_ref(rm_inside_link(rm_emph(kiso))))))
good_tmpl_dict = {key: value for key, value in re.findall(tmpl_pattern, good_text)}
pprint(good_tmpl_dict)
````

29. 国旗画像のURLを取得する
テンプレートの内容を利用し，国旗画像のURLを取得せよ．（ヒント: MediaWiki APIのimageinfoを呼び出して，ファイル参照をURLに変換すればよい）


````python
import urllib3

flag_file = good_tmpl_dict['国旗画像']
wiki_base_url = 'https://www.mediawiki.org/w/api.php?'
paramaters = {'action': 'query', # query投げる
              'format': 'json', # jsonで返す
              'titles': 'File:' + flag_file.replace(' ', '_'),
              'prop': 'imageinfo', # imageinfo に設定
              'iiprop': 'url' # url を取ってくる
              }
# paramater の情報ををurlに埋め込み
paramaters_url = '&'.join(f'{item[0]}={item[1]}' for item in paramaters.items())
http = urllib3.PoolManager()
# request投げる
r = http.request('GET', wiki_base_url + paramaters_url)
data = json.loads(r.data)
url = data['query']['pages']['-1']['imageinfo'][0]['url']
print(url)
````
